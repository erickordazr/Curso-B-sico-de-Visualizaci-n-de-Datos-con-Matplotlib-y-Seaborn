Diferentes gráficos que ofrece Seaborn para graficar datos categóricos, variables de texto

Datos

```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

Gráfica de cantidad


```python
sns.countplot(data=tips,x='day',hue='sex')
plt.show()
```

Diagrama de puntos


```python
plt.figure(figsize=(6,6))
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

Diagrama de caja
(Se pueden combinar distintos gráficos)

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,color="0",marker='<')
plt.show()
```

Violín: parecido a Boxplot, pero no muestra los cuartiles, sino como se concentran los datos con comparación split


```python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',split=True,palette='pastel')
plt.show()
```

```python
sns.violinplot(data=tips,y='total_bill')
```

Catplot, con kind podemos cambiar el tipo de gráfico y se puede agregar otra variable col


```python
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,kind='swarm',col='time')
plt.show()
```

```python
sns.catplot(x="day", hue="smoker",
            col="time", 
            kind="count", data=tips)

plt.show()
```
